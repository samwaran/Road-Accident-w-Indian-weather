### Phase 4: Visualising the Data

##### 4.1 States/UTs where persons Killed on a sunny weather is greater than 5000 across 2018 

```python
filt = df['Sunny_person_killed_N']>5000
Plot = df.loc[filt, ('States/UTs', 'Sunny_person_killed_N')].sort_values(by='Sunny_person_killed_N',ascending=False)

plt.figure(figsize =(15, 7))

plt.bar(x=Plot['States/UTs'],height=Plot['Sunny_person_killed_N'], width=0.4)

plt.show()
```

**Key Takeaway:** There are 11 States/UTs where persons Killed on a sunny weather is greater than 5000 across 2018.

**Mahrashtra being the highest, & kerala having the least amount of cases.**

##### 4.2 Visulisation on 10 States/UTs with least accidents on sunny days and it's correlation analysis with accidents on all weather conditions. 

```python
Plot = df[['States/UTs', 'Sunny_total_acc_N', 'Rain_total_acc_N', 'Fog_total_acc_N', 'Hail_total_acc_N']].sort_values(by='Sunny_total_acc_N', ascending=True).head(10)
#Plot = df.loc[filt, ('States/UTs', 'Sunny_person_killed_N', 'Rain_person_killed_N')].sort_values(by='Sunny_person_killed_N',ascending=False)
barWidth = 0.2
plt.figure(figsize =(20, 7))

br1 = np.arange(len(Plot['Sunny_total_acc_N']))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

plt.bar(x=br1,height=Plot['Sunny_total_acc_N'], label ='Accidents on Sunny', width=barWidth, color='orange')
plt.bar(x=br2,height=Plot['Rain_total_acc_N'], label ='Accidents on Rainy', width=barWidth, color='blue')
plt.bar(x=br3,height=Plot['Fog_total_acc_N'], label ='Accidents on Foggy', width=barWidth, color='red')
plt.bar(x=br4,height=Plot['Hail_total_acc_N'], label ='Accidents on Hail', width=barWidth, color='green')
plt.xticks(ticks=(br1+br2+br3+br4)/4, labels=Plot['States/UTs'])

plt.legend()
plt.show()
```

**Key Takeaways:** Lakshdweep Having the least cases during all weather.<br/>
Arunachal pradesh having morderate cases in all weathers may suggest infrastructural problems. 

*Though it needs further investigation on secondary lines for concrete analysis.*

##### 4.3 Distribution of deaths in Uttar Pradesh under different weather conditions

```python
filt = df['States/UTs'] == 'Uttar Pradesh'
deaths_sunny = int(df.loc[filt, 'Sunny_person_killed_N'])
deaths_rainy = int(df.loc[filt, 'Rain_person_killed_N'])
deaths_fog = int(df.loc[filt, 'Fog_person_killed_N'])
deaths_hail = int(df.loc[filt, 'Hail_person_killed_N'])
sum_deaths = deaths_sunny + deaths_rainy + deaths_fog + deaths_hail
data=[deaths_sunny/sum_deaths*100, deaths_rainy/sum_deaths*100, deaths_fog/sum_deaths*100, deaths_hail/sum_deaths*100]

plt.pie(data, labels=['Sun', 'Rain', 'Fog', 'Hail'], autopct='%1.1f%%')
plt.title('Deaths in Uttar Pradesh under different weather conditions')

plt.show()
```

**Key Takeaways:** The accidents are higher in sunny day indicating rash driving or careless traffic system.<br/>

*Would require further investigation for concrete analysis*
